# Uber - Data Modelling & Model Evaluation (Hourly Traffic Forecasting)

This notebook trains and evaluates models that forecast **hourly traffic volumes** at each junction, using
calendar features and special events (and weather if present). It follows a time-aware validation protocol.

**Metrics:** MAE, RMSE, R².  
**CV:** TimeSeriesSplit (rolling-origin).  
**Models:** Baselines (naïve, seasonal), ARIMA/SARIMAX (if available), tree-based (Gradient Boosting, Random Forest), optional LSTM (if TensorFlow available).

**References to keep in mind:**

- Towards Data Science — *Time Series Forecasting with Machine Learning*.  https://towardsdatascience.com/time-series-forecasting-with-machine-learning-b3072a5b44ba/

```python
# 1) Load integrated dataset & build features
df = pd.read_csv(PATHS["integrated_csv"])
print("Rows x Cols:", df.shape)

# Column detection
time_col = "timestamp_hour" if "timestamp_hour" in df.columns else [c for c in df.columns if "time" in c.lower()][0]
junction_col = "Junction" if "Junction" in df.columns else [c for c in df.columns if "junction" in c.lower()][0]
target_col = "vehicles" if "vehicles" in df.columns else [c for c in df.columns if "vehicle" in c.lower()][0]

# Parse time
df[time_col] = pd.to_datetime(df[time_col], errors="coerce")
df = df.dropna(subset=[time_col, target_col, junction_col]).copy()
df = df.sort_values([junction_col, time_col]).reset_index(drop=True)

# Calendar features
df["year"] = df[time_col].dt.year
df["month"] = df[time_col].dt.month
df["day"] = df[time_col].dt.day
df["hour"] = df[time_col].dt.hour
df["dow"] = df[time_col].dt.dayofweek
df["is_weekend"] = (df["dow"] >= 5).astype(int)

# Event features (present in integrated dataset)
event_like = [c for c in df.columns if c.startswith("evt_")]
holiday_cols = [c for c in df.columns if c in ["is_public_holiday", "is_any_event", "event_count"]]

# Weather (optional)
weather_like = [c for c in ["temp_c","rh_pct","precip_mm","wind_speed_ms","fog_risk","is_rain","is_thunder",
                            "rain_heavy","rain_mod","rain_light","wind_strong","wind_extreme",
                            "precip_mm_3h","precip_mm_6h","temp_c_lag1","precip_lag1"]
                if c in df.columns]

# Lag features (built from target per junction to avoid leakage across sites)
def add_lags_rolls(g):
    g = g.sort_values(time_col).copy()
    for L in [1, 24, 168]:
        g[f"lag_{L}h"] = g[target_col].shift(L)
    g["roll_mean_3h"] = g[target_col].rolling(3, min_periods=1).mean()
    g["roll_mean_24h"] = g[target_col].rolling(24, min_periods=1).mean()
    g["roll_std_24h"] = g[target_col].rolling(24, min_periods=2).std()
    return g

df = df.groupby(junction_col, group_keys=False).apply(add_lags_rolls)
df = df.dropna(subset=["lag_1h","lag_24h","lag_168h"]).reset_index(drop=True)

# Feature list
base_feats = ["year","month","day","hour","dow","is_weekend"]
X_cols = base_feats + event_like + holiday_cols + weather_like +          ["lag_1h","lag_24h","lag_168h","roll_mean_3h","roll_mean_24h","roll_std_24h"]

# Keep only columns that exist
X_cols = [c for c in X_cols if c in df.columns]
print("Using features:", len(X_cols))

# Train/holdout split by time (last 15% of timestamps as holdout)
time_cut = df[time_col].quantile(0.85)
train_df = df[df[time_col] <= time_cut].copy()
holdout_df = df[df[time_col] > time_cut].copy()

print("Train range:", train_df[time_col].min(), "→", train_df[time_col].max(), "| Rows:", len(train_df))
print("Holdout range:", holdout_df[time_col].min(), "→", holdout_df[time_col].max(), "| Rows:", len(holdout_df))
```

```python
# 2) Baseline forecasts (naïve & seasonal)
def mae(y_true, y_pred): return mean_absolute_error(y_true, y_pred)
def rmse(y_true, y_pred): return mean_squared_error(y_true, y_pred, squared=False)
def r2(y_true, y_pred): return r2_score(y_true, y_pred)

def eval_baseline(df_in, horizon_mask, lags):
    """Evaluate baselines using specified lag hours on the provided mask.
    lags: list of lag integers, e.g., [1, 24, 168] for naïve-1h, seasonal day, seasonal week.
    """
    out = []
    for L in lags:
        df_in[f"pred_lag_{L}h"] = df_in.groupby(junction_col)[target_col].shift(L)
        sub = df_in.loc[horizon_mask & df_in[f"pred_lag_{L}h"].notna(), [target_col, f"pred_lag_{L}h"]]
        if len(sub) == 0:
            continue
        m_mae = mae(sub[target_col], sub[f"pred_lag_{L}h"])
        m_rmse = rmse(sub[target_col], sub[f"pred_lag_{L}h"])
        m_r2 = r2(sub[target_col], sub[f"pred_lag_{L}h"])
        out.append({"model":"baseline_lag", "variant":f"L={L}h", "MAE":m_mae, "RMSE":m_rmse, "R2":m_r2})
    return pd.DataFrame(out)

# Evaluate baselines on holdout only
holdout_mask = df[time_col] > time_cut
baseline_metrics = eval_baseline(df, holdout_mask, [1, 24, 168])
baseline_metrics
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import TimeSeriesSplit
from sklearn.ensemble import GradientBoostingRegressor, RandomForestRegressor
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_absolute_error as mae, mean_squared_error, r2_score as r2

# Define RMSE separately
def rmse(y_true, y_pred):
    return mean_squared_error(y_true, y_pred, squared=False)

# ---- Inputs ----
# Assumes these are defined:
# train_df, holdout_df, X_cols, target_col, junction_col

# Copy base data
train_proc = train_df.copy()
holdout_proc = holdout_df.copy()

# Convert boolean columns to numeric (if any)
for df in [train_proc, holdout_proc]:
    bool_cols = df.select_dtypes(include=["bool"]).columns
    for col in bool_cols:
        df[col] = df[col].astype(int)

# One-hot encode junction
train_proc = pd.concat([train_proc, pd.get_dummies(train_proc[junction_col], prefix="J")], axis=1)
holdout_proc = pd.concat([holdout_proc, pd.get_dummies(holdout_proc[junction_col], prefix="J")], axis=1)

# Align dummy columns between train and holdout
dummy_cols = [col for col in train_proc.columns if col.startswith("J_")]
for col in dummy_cols:
    if col not in holdout_proc.columns:
        holdout_proc[col] = 0
holdout_proc = holdout_proc.reindex(columns=train_proc.columns, fill_value=0)

# Prepare numeric feature list (exclude original junction_col and any non-numeric)
numeric_feature_cols = [
    col for col in X_cols
    if col != junction_col and pd.api.types.is_numeric_dtype(train_proc[col])
]

# Impute and scale numeric features
num_imputer = SimpleImputer(strategy="median")
scaler = StandardScaler()

X_num_train = num_imputer.fit_transform(train_proc[numeric_feature_cols])
X_num_train = scaler.fit_transform(X_num_train)

X_num_hold = num_imputer.transform(holdout_proc[numeric_feature_cols])
X_num_hold = scaler.transform(X_num_hold)

# Combine numeric and dummy features
X_train = np.hstack([X_num_train, train_proc[dummy_cols].values])
X_hold  = np.hstack([X_num_hold,  holdout_proc[dummy_cols].values])

# Targets
y_train = train_proc[target_col].values
y_hold  = holdout_proc[target_col].values

# TimeSeriesSplit
tscv = TimeSeriesSplit(n_splits=5)

# Evaluation function
def eval_sklearn_model(model, X_tr, y_tr, X_te, y_te, name):
    cv_mae = []; cv_rmse = []; cv_r2 = []

    for tr_idx, va_idx in tscv.split(X_tr):
        X_tr_i, X_va_i = X_tr[tr_idx], X_tr[va_idx]
        y_tr_i, y_va_i = y_tr[tr_idx], y_tr[va_idx]
        model.fit(X_tr_i, y_tr_i)
        p = model.predict(X_va_i)
        cv_mae.append(mae(y_va_i, p))
        cv_rmse.append(rmse(y_va_i, p))
        cv_r2.append(r2(y_va_i, p))

    model.fit(X_tr, y_tr)
    p_hold = model.predict(X_te)

    return {
        "model": name,
        "CV_MAE_mean": float(np.mean(cv_mae)), "CV_MAE_std": float(np.std(cv_mae)),
        "CV_RMSE_mean": float(np.mean(cv_rmse)), "CV_RMSE_std": float(np.std(cv_rmse)),
        "CV_R2_mean": float(np.mean(cv_r2)), "CV_R2_std": float(np.std(cv_r2)),
        "HOLDOUT_MAE": mae(y_te, p_hold),
        "HOLDOUT_RMSE": rmse(y_te, p_hold),
        "HOLDOUT_R2": r2(y_te, p_hold)
    }, p_hold

# ---- Train Models ----
results = []
pred_store = {}

# Gradient Boosting
gbr = GradientBoostingRegressor(random_state=42)
res_gbr, pred_gbr = eval_sklearn_model(gbr, X_train, y_train, X_hold, y_hold, "GradientBoosting")
results.append(res_gbr)
pred_store["GradientBoosting"] = pred_gbr

# Random Forest
rf = RandomForestRegressor(n_estimators=300, max_depth=None, min_samples_leaf=5, random_state=42, n_jobs=-1)
res_rf, pred_rf = eval_sklearn_model(rf, X_train, y_train, X_hold, y_hold, "RandomForest")
results.append(res_rf)
pred_store["RandomForest"] = pred_rf

# Results DataFrame
cv_df = pd.DataFrame(results)
print(cv_df)
```

```python
# 4) Hyperparameter tuning (RandomizedSearch on GradientBoosting with TimeSeriesSplit)
param_dist = {
    "n_estimators": [200, 300, 400],
    "max_depth": [2, 3, 4, 5],
    "learning_rate": [0.05, 0.075, 0.1, 0.2],
    "subsample": [0.6, 0.8, 1.0],
    "min_samples_leaf": [1, 3, 5, 10]
}
gbr0 = GradientBoostingRegressor(random_state=42)
search = RandomizedSearchCV(gbr0, param_distributions=param_dist, n_iter=15, scoring="neg_mean_absolute_error",
                            cv=tscv, random_state=42, n_jobs=-1, verbose=1)
search.fit(X_train, y_train)

best_gbr = search.best_estimator_
res_gbr_tuned, pred_gbr_tuned = eval_sklearn_model(best_gbr, X_train, y_train, X_hold, y_hold, "GradientBoosting[Tuned]")
cv_df = pd.concat([cv_df, pd.DataFrame([res_gbr_tuned])], ignore_index=True)
pred_store["GradientBoosting[Tuned]"] = pred_gbr_tuned

search.best_params_, res_gbr_tuned
```

```python
# 5) Optional: SARIMAX per junction (with exogenous features if available)

    # We'll fit a simple SARIMAX per junction with weekly seasonality, using exogenous regressors (features)
    # To keep runtime manageable, we aggregate to a single global model per junction and predict holdout.
print("SARIMAX block is running.")
print("HAS_STATSMODELS =", HAS_STATSMODELS)

if HAS_STATSMODELS:
    sarimax_metrics = []
    sarimax_preds = []

    exog_cols = [c for c in X_cols if not c.startswith("lag_")]
    print("Exogenous columns:", exog_cols)

    for j_id, g in df.groupby(junction_col):
        g = g.sort_values(time_col).copy()
        g_tr = g[g[time_col] <= time_cut]
        g_ho = g[g[time_col] > time_cut]

        print(f"Junction: {j_id} | Train: {len(g_tr)} | Holdout: {len(g_ho)}")
        if len(g_tr) < 250 or len(g_ho) == 0:
            print(f"Skipping junction {j_id} due to insufficient data.")
            continue

        order = (1, 0, 1)
        seasonal_order = (1, 1, 1, 24)

        try:
            print(f"Fitting SARIMAX for junction {j_id}...")
            model = SARIMAX(
                g_tr[target_col],
                order=order,
                seasonal_order=seasonal_order,
                exog=g_tr[exog_cols] if exog_cols else None,
                enforce_stationarity=False,
                enforce_invertibility=False
            )
            res = model.fit(disp=False)
            fc = res.predict(start=g_ho.index[0], end=g_ho.index[-1],
                             exog=g_ho[exog_cols] if exog_cols else None)
            m_mae = mae(g_ho[target_col].values, fc.values)
            m_rmse = rmse(g_ho[target_col].values, fc.values)
            m_r2 = r2(g_ho[target_col].values, fc.values)
            sarimax_metrics.append({
                "model": "SARIMAX", "junction": j_id,
                "HOLDOUT_MAE": m_mae, "HOLDOUT_RMSE": m_rmse, "HOLDOUT_R2": m_r2
            })
            sarimax_preds.append(pd.DataFrame({
                junction_col: j_id,
                time_col: g_ho[time_col].values,
                "y_true": g_ho[target_col].values,
                "y_pred": fc.values
            }))
            print(f"SARIMAX done for junction {j_id} | MAE: {m_mae:.2f}")
        except Exception as e:
            print("SARIMAX failed for junction", j_id, ":", str(e))

    if sarimax_metrics:
        sarimax_df = pd.DataFrame(sarimax_metrics)
        display(sarimax_df.head())
        if 'cv_df' not in locals():
            cv_df = pd.DataFrame()
        cv_df = pd.concat([cv_df, sarimax_df.groupby("model")[["HOLDOUT_MAE", "HOLDOUT_RMSE", "HOLDOUT_R2"]].mean().reset_index()], ignore_index=True)
        sarimax_pred_df = pd.concat(sarimax_preds, ignore_index=True)
        print("SARIMAX results stored.")
    else:
        print("No SARIMAX models were successfully fitted.")
else:
    print("statsmodels not available; skipping SARIMAX.")
```

```python
# 6) Optional: Simple LSTM (guarded)
    # Prepare a single global LSTM across junctions. We'll use lag features as supervised inputs.
    # Shape: (samples, timesteps, features). We'll assemble a short sliding window.
if HAS_TF:
    import numpy as np
    import pandas as pd
    import tensorflow as tf
    from tensorflow import keras
    from sklearn.preprocessing import StandardScaler
    from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

    # ---------------- Params ----------------
    window = 24                  # lookback steps
    feat_cols = X_cols           # your feature columns
    rng_seed = 42

    # ---------------- Determinism ----------------
    np.random.seed(rng_seed)
    tf.random.set_seed(rng_seed)

    # ---------------- Helpers ----------------
    def to_numeric_col(s: pd.Series) -> pd.Series:
        """Make feature column numeric:
           - object -> binary yes/no mapping
           - bool -> 0/1
           - numeric -> float
           - otherwise -> coerce to numeric
        """
        if s.dtype == "object":
            # anything not an explicit 'no/false/0' becomes 1 (if not NA)
            def map_obj(x):
                if pd.isna(x):
                    return np.nan
                v = str(x).strip().lower()
                return 0.0 if v in {"no", "false", "0", "nan"} else 1.0
            return s.apply(map_obj).astype(float)

        if pd.api.types.is_bool_dtype(s):
            return s.astype(float)

        if pd.api.types.is_numeric_dtype(s):
            return s.astype(float)

        # last resort
        return pd.to_numeric(s, errors="coerce").astype(float)

    def make_sequences(df: pd.DataFrame, scaler: StandardScaler | None, fit_scaler: bool):
        """Sort by time, select cols, convert to numeric, drop NA, scale (train-only fit),
        then build (window, features) sequences."""
        df = df.sort_values(time_col).copy()
        df = df[feat_cols + [target_col]]

        # convert feature columns to numeric/binary
        for c in feat_cols:
            df[c] = to_numeric_col(df[c])

        # drop rows with NA in any used column (features + target)
        df = df.dropna()

        X = df[feat_cols].values.astype(np.float32)
        y = df[target_col].values.astype(np.float32)

        if fit_scaler:
            scaler = StandardScaler()
            X = scaler.fit_transform(X).astype(np.float32)
        else:
            if scaler is None:
                raise ValueError("Scaler must be provided when fit_scaler=False")
            X = scaler.transform(X).astype(np.float32)

        # windowed sequences
        X_seq, y_seq = [], []
        for i in range(window, len(X)):
            X_seq.append(X[i - window:i])
            y_seq.append(y[i])

        X_seq = np.asarray(X_seq, dtype=np.float32)
        y_seq = np.asarray(y_seq, dtype=np.float32)

        # sanity checks
        assert np.isfinite(X_seq).all(), "X_seq contains NaNs or Infs"
        assert np.isfinite(y_seq).all(), "y_seq contains NaNs or Infs"

        return X_seq, y_seq, scaler, df  # return df for naive baseline if needed

    # ---------------- Train sequences (+ time-aware val split) ----------------
    X_all, y_all, scaler, df_train_clean = make_sequences(train_df, scaler=None, fit_scaler=True)

    # take a small tail as validation to respect temporal order
    val_frac = 0.1
    n_val = max(1, int(len(X_all) * val_frac))
    X_train, y_train = X_all[:-n_val], y_all[:-n_val]
    X_val, y_val = X_all[-n_val:], y_all[-n_val:]

    # ---------------- Model ----------------
    model = keras.Sequential([
        keras.layers.Input(shape=(window, len(feat_cols))),
        keras.layers.LSTM(64),
        keras.layers.Dense(32, activation="relu"),
        keras.layers.Dense(1),
    ])
    model.compile(optimizer=keras.optimizers.Adam(1e-3), loss="mae", metrics=["mae"])

    callbacks = [
        keras.callbacks.EarlyStopping(monitor="val_mae", patience=5, restore_best_weights=True),
        keras.callbacks.ReduceLROnPlateau(monitor="val_mae", factor=0.5, patience=3, min_lr=1e-5),
    ]

    model.fit(
        X_train, y_train,
        validation_data=(X_val, y_val),
        epochs=50,
        batch_size=128,
        shuffle=False,   # VERY important for time series
        verbose=1,
        callbacks=callbacks
    )

    # ---------------- Holdout ----------------
    X_ho, y_ho, _, df_ho_clean = make_sequences(holdout_df, scaler=scaler, fit_scaler=False)

    if X_ho.size:
        yhat_ho = model.predict(X_ho, verbose=0).ravel()

        mae = mean_absolute_error(y_ho, yhat_ho)
        rmse = mean_squared_error(y_ho, yhat_ho, squared=False)
        r2 = r2_score(y_ho, yhat_ho)

        # Naive "persistence" baseline: y_t ≈ y_{t-1}
        tgt_vec = df_ho_clean[target_col].values.astype(np.float32)
        naive_pred = tgt_vec[window-1:-1]  # aligns with y_ho = tgt_vec[window:]
        naive_mae = mean_absolute_error(y_ho, naive_pred)

        print(f"LSTM holdout  MAE: {mae:.4f} | RMSE: {rmse:.4f} | R²: {r2:.4f}")
        print(f"Naive baseline MAE: {naive_mae:.4f}")
    else:
        print("Not enough holdout data after windowing.")
else:
    print("TensorFlow/Keras not available; skipping LSTM.")
```
